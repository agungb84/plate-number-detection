```python
import google.generativeai as genai
from PIL import Image
import matplotlib.pyplot as plt

# Initialize Gemini model
model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest")

# Load image
sample_file_3 = Image.open('/content/plat3.jfif')

# Display the image
plt.imshow(sample_file_3)
plt.axis('off')  # Menyembunyikan axis untuk tampilan gambar yang lebih bersih
plt.show()

# Create prompt
prompt = "Read the plate number of the image with the output example { 'plate_no': 'CD 7528A', 'vehicle': 'car', 'vehicle_type': 'MPV', 'color': 'white', 'gate_open': 'N/A', 'gate_closed': 'N/A' }"

# Generate response
response = model.generate_content([sample_file_3, prompt])

# Print response
print(response.text)
```
